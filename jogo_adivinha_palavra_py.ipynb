```python
import getpass


palavra_secreta = str(getpass.getpass('digite uma palavra: '))
print('>' * 30)
print(len(palavra_secreta))
letras_acertadas = ''
contador = 0
palavra_formada = ''


while True:

    palavra_digitada = str(input('Digite uma letra: '))
    if palavra_digitada == '10':
      break



    if len(palavra_digitada) >=2 or len(palavra_digitada) == 0:
      print('''voçe digitou mais de um caractere
      tente novamente......''')
    else:


      contador += 1
      # verificando se a letra digitada esta na palavra secreta e criando variavel so com as palavras acertadas
      if palavra_digitada in palavra_secreta:
        letras_acertadas += palavra_digitada

      # mostrado as letras na palavra ou exibir * se nao estiver localizado
      for letras in palavra_secreta:
        if letras in letras_acertadas:
          #adicionando as letras em uma variavem obs. so adiciona se nao estiver
          if letras not in palavra_formada:
            palavra_formada += letras

          print(letras, end='')
        else:
          print('*', end='')
      print()

    if palavra_secreta == palavra_formada:
      print(f'parabensssss  voce ganhou a palavra era: {palavra_secreta}')
      print(f'voce fez {contador} tentativas.... ')
      break











```
