```python
def jogo_pergunta_respostas():

  perguntas = [
        {
            'pergunta':'quanto e 2 + 2?',
            'opcoes': ['1','2','3','4'],
            'resposta': '4',
        },
        {
            'pergunta':'quanto e 5 + 5?',
            'opcoes': ['25','55','10','51'],
            'resposta': '10',
        },
        {
            'pergunta':'quanto e 10 / 2?',
            'opcoes': ['4','5','20','11'],
            'resposta': '5',
        },
  ]
  contador_perguntas = 0
  contador_acertos = []

  while contador_perguntas <= 2:
    def verifica_pergunta():

        #exibindo a pergunta
      print('pergunta: ',perguntas[contador_perguntas - 1]['pergunta'])
      contador = 0

      #mostrando as opçoes
      for opcoes_1 in perguntas[contador_perguntas - 1]['opcoes']:
        print(f'({contador})', opcoes_1)
        contador +=1

      #esse while serve para nao digitar nada diferente das opçoes
      while True:
        try:
          escolha_de_opcao = int(input('escolha uma opção: '))
          if escolha_de_opcao <=3:
            break
          else:
            print('opção invalida')
        except ValueError:
          print('opção invalida')

        #colocando a resposta em uma variavel aterando junto com a pergunta
      resposta_correta = perguntas[contador_perguntas - 1]['resposta']

      checagem_de_opcao = perguntas[contador_perguntas - 1]['opcoes'][escolha_de_opcao]

        # vericando se esta certo ou errado a opçao escolhida
      if checagem_de_opcao == resposta_correta:
        contador_acertos.append(1)
        print('..... Parabens voçê acerto..... ',("\U0001F600")*3)
      else:
        print('''
        .... resposta errada ....
        .... tente novamente ....''')
      print('--' * 30)

    contador_perguntas +=1
    #execulta a funçao que esta dentro da funçao
    chama_funçao = verifica_pergunta()

  if len(contador_acertos) > 0:
    print(f'''
    >>>>>>>👏👏  PARABENS 👏👏<<<<<<<<
    voce acertor {len(contador_acertos)} perguntas''')
  else:
    print('😢😢 infelizmente voce erro todas respostas')


jogo_pergunta_respostas()
```
